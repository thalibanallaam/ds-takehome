```python
import pandas as pd

# Load the dataset
file_path = '../data/e_commerce_transactions.csv'
df = pd.read_csv(file_path)

# Display basic info and preview of the "decoy_noise" feature
df['decoy_noise'].describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set up visualizations
plt.figure(figsize=(14, 6))

# Histogram
plt.subplot(1, 2, 1)
sns.histplot(df['decoy_noise'], bins=50, kde=True)
plt.title('Histogram of decoy_noise')

# Boxplot
plt.subplot(1, 2, 2)
sns.boxplot(x=df['decoy_noise'])
plt.title('Boxplot of decoy_noise')

plt.tight_layout()
plt.show()
```
